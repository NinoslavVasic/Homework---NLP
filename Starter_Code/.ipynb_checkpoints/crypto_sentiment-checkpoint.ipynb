# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Read your api key environment variable

api_key = os.getenv("news_api")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
BTC_headlines = newsapi.get_everything(q="bitcoin AND Bitcoin AND BTC", page_size=50,language="en", sort_by="relevancy")

```

```python
# Print all publshed articles about Bitcoin

print(f"Total number of articles about Bitcoin: {BTC_headlines['totalResults']}")
# Show sample article
BTC_headlines["articles"][0]
```

```python
# Fetch the Ethereum news articles

ETH_headlines = newsapi.get_everything(q="ethereum AND Ethereum AND ETH",language="en",page_size=50,sort_by="relevancy")

# Print total articles
print(f"Total number of articles about Ethereum: {ETH_headlines['totalResults']}")
# Show sample article
ETH_headlines["articles"][0]
```

```python
# Create the Bitcoin sentiment scores DataFrame
BTC_sentiments = []

for btc_article in BTC_headlines["articles"]:
    try:
        text = btc_article["content"].lower()    
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        BTC_sentiments.append({
            "Text": text,
            "Compound": compound,
            "Positive": pos,
            "Negative": neg,
            "Neutral": neu,  
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
BTC_df = pd.DataFrame(BTC_sentiments)

# Reorder DataFrame columns
cols = ["Compound","Negative","Neutral","Positive","Text"]
BTC_df = BTC_df[cols]

BTC_df.head()
```

```python
# Create the ethereum sentiment scores DataFrame

ETH_sentiments = []
for eth_article in ETH_headlines["articles"]:
    try:
        text = eth_article["content"].lower()
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        ETH_sentiments.append({
            "Text": text,
            "Compound": compound,
            "Positive": pos,
            "Negative": neg,
            "Neutral": neu,
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
ETH_df = pd.DataFrame(ETH_sentiments)

# Reorder DataFrame columns
cols = ["Compound","Negative","Neutral","Positive","Text"]
ETH_df = ETH_df[cols]


ETH_df.head(10)
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary
stop_words_ = {'damn','char','much','ha', 'fucking'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    # Create a list of the words
    sw = set(stopwords.words('english'))
     # Remove the non-alpha characters # Substitute everything that is NOT a letter with empty string
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    words = word_tokenize(re_clean)
   
    # Lemmatize Words into root words
    lemmatizer = WordNetLemmatizer() 
    lem = [lemmatizer.lemmatize(word) for word in words]
    
    #convert to lower case and remove the stopwords 
    tokens = [word.lower() for word in lem if word.lower() not in sw.union(stop_words_)]
    
    return tokens
        
```

```python
# Create a new tokens column for bitcoin

BTC_tokens = BTC_df.copy()
tokenized = []
for i in BTC_tokens['Text']:
    tokenized_articles = tokenizer(i)
    tokenized.append({'tokens':tokenized_articles})

BTC_tokens_df= pd.DataFrame(tokenized)
BTC_tokens['Tokens'] = BTC_tokens_df 


BTC_tokens.head()
```

```python
# Create a new tokens column for ethereum
ETH_tokens = ETH_df.copy()
tokenized = []
for i in ETH_tokens['Text']:
    tokenized_articles = tokenizer(i)
    tokenized.append({'tokens':tokenized_articles})

ETH_tokens_df= pd.DataFrame(tokenized)
ETH_tokens['Tokens'] = ETH_tokens_df 
  
ETH_tokens.tail()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
BTC_words_all = []
for text in BTC_tokens['Tokens']:
    for word in text:
        BTC_words_all.append(word)
BTC_word_counts = Counter(ngrams(BTC_words_all, n=2))
BTC_word_counts.most_common(10)
```

```python
# Generate the Ethereum N-grams where N=2
ETH_words_all = []
for text in ETH_tokens['Tokens']:
    for word in text:
        ETH_words_all.append(word)
ETH_word_counts = Counter(ngrams(ETH_words_all, n=2))
ETH_word_counts.most_common(10)
```

```python
# Use the token_count function to generate the top 10 words from each coin
tokens = [BTC_tokens, ETH_tokens]
def token_count(tokens, N=10):
    
    """Returns the top N tokens from the frequency count"""
    
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin

token_count(BTC_words_all)
```

```python
# Get the top 10 words for Ethereum
token_count(ETH_words_all)
```

# Word Clouds

In this section, you will generate word clouds for each coin to summarize the news for each coin

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [20.0, 10.0]
```

```python
#Set the function
def processed_text(corpus): 
    # Combine all articles in corpus into one large string
    big_string = ' '.join(corpus)
    return big_string
BTC_words = processed_text(BTC_words_all)
ETH_words = processed_text(ETH_words_all)
```

```python
# Generate the Bitcoin word cloud
BTC_wc = WordCloud(collocations=False).generate(BTC_words)
fig = plt.figure()
plt.imshow(BTC_wc)
plt.title('Word Cloud - BITCOIN (BTC)', fontsize=30, fontweight='bold')
plt.axis('off')
plt.show()
```

```python
# Generate the Ethereum word cloud

fig = plt.figure()
ETH_wc = WordCloud().generate(ETH_words)
plt.imshow(ETH_wc)
plt.title('Word Cloud - ETHEREUM (ETH)', fontsize=30, fontweight='bold')
plt.axis('off')
plt.show()
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
#!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
BTC_sent = []
for sent in BTC_tokens['Text']:
    sent_list = sent
    BTC_sent.append(sent_list)
    
one_string_BTC = ' '.join(BTC_sent)
one_string_BTC
```

```python
# Run the NER processor on all of the text
BTC_doc = nlp(one_string_BTC)

# Add a title to the document
BTC_doc.user_data['title'] = 'Bitcoin (BTC) NER'
```

```python
# Render the visualization
displacy.render(BTC_doc, style='ent')
```

```python
# List all Entities
for ent in BTC_doc.ents:
    print (ent, ent.label_)
```
